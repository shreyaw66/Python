```python
import pandas as pd
from datasets import load_dataset
import ast
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading data

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
#datacleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#filter for US data analytic role

df_US_data_roles = df[(df['job_country'] == "United States") & (df['job_title_short']=='Data Analyst')]

df_US_data_roles
```

```python
#horizontal bar graph for plotting job count vs location
df_plot = df_US_data_roles['job_location'].value_counts().head(10).to_frame()

sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of jobs per Job title')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```

```python
#subplot to showcase piechart 

dict_column = {
        'job_work_from_home': 'Work from home Offered',
        'job_no_degree_mention':'Degree Requirement',
        'job_health_insurance': 'Health Insurance Offered'
}

fig, ax =plt.subplots(1,3)
fig.set_size_inches((12,5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_US_data_roles[column].value_counts(), labels = ['False','True'], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

plt.show()
```

```python
#horizontal bar graph for plotting job count vs company name
df_plot = df_US_data_roles['company_name'].value_counts().head(10).to_frame()

sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Companies for Data Analyst role in US')
plt.xlabel('Number of jobs')
plt.ylabel('')
plt.show()
```
